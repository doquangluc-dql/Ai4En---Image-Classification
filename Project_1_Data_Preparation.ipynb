## Tải dữ liệu

Tải dữ liệu đã xử lý sẵn trên Google Drive

```python
import os, gdown
import zipfile
project_dir = 'Project 1 - CV - Animal Image Classification/'
filter_path = project_dir + 'filtered-name/'
image_path = project_dir + 'animals/'
os.makedirs(image_path, exist_ok=True)

image_dataset_url = 'https://drive.google.com/file/d/19Rr-b09YUjcmgc6_mJyTZHSOsr-ZPNNP/view?usp=drive_link'
zipfile_path = os.path.join(project_dir, 'animals.zip')
gdown.download(image_dataset_url, zipfile_path, quiet=False, fuzzy=True)
with zipfile.ZipFile(zipfile_path, 'r') as zip_ref:
    zip_ref.extractall(project_dir)
```

```python
for folder in folders:
  file_path = f'{filter_path}/{folder}/filtered-name.txt'

  with open(file_path, 'r') as file:
    for line in file:
      img = line.strip()
      path_img = f'{image_path}/{folder}/{img}'
      if os.path.exists(path_img):
        os.remove(path_img)
      else:
        print(f"The file {path_img} does not exist.")
```
